### EDA with complete dataset

```python
# libraries
import os
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set path
parent_directory = os.path.abspath(os.path.join('..'))
PATH = os.path.join(parent_directory, 'data')

# helper functions
def load_data(filename):
    '''load the data with only filename'''
    return pd.read_csv(os.path.join(PATH, filename))

def get_distribution(df, cols, figsize, fontsize=8):
    '''get distribution of data'''
    plt.figure(figsize=figsize)
    for i, col in enumerate(cols):
        plt.subplot(8, 3, i+1)  # Create subplots for each numerical column
        sns.histplot(df[col], bins=30, kde=True)  # Histogram with KDE for better visualization
        plt.title(f'Distribution of {col}', fontsize=fontsize)
        plt.xlabel(col, fontsize=fontsize)
        plt.ylabel('Frequency', fontsize=fontsize)

    plt.tight_layout()
    plt.show()
```

```python
# load updated dataset with all columns
df = load_data('cleaned_BankChurner_as_in_16_sept.csv')
```

```python
# get distribution of the data
numerical_columns = [
    'Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count',
    'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
    'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct',
    'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio'
] 

get_distribution(df, numerical_columns, figsize=(10, 8))
```
